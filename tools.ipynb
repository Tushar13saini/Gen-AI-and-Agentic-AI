```python
from dotenv import load_dotenv  
load_dotenv()  
```

```python
from langchain.agents import create_agent
```

```python
def first_agent(city:str)-> str:
    """used to get the weather information for the given city"""
    return f"The current weather in {city} is very cloudy."
```

```python
first_agent("panipat")
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="llama-3.3-70b-versatile")
```

```python
agent = create_agent(
    model=llm,
    tools=[first_agent],
    system_prompt="You are a helpful assistant that provides weather information."
)
```

```python
weather = input("enter the city name to get the weather information: ")
```

```python
response = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": f"What is the weather in {weather}?"
        }
    ]
})
```

```python
response["messages"]
```

```python
from langchain.chat_models import init_chat_model 
```

```python
model = init_chat_model("groq:llama-3.3-70b-versatile") 
```

```python
model.invoke("what is the weather today?") 
```

```python
#batch processing
response =model.batch(["what is programming?",
                       "who is bill gates?",
                       "what is the weather today?",
                       "what is the capital of india?"],
                config={
                    "max_concurrency": 2 #make our work faster by processing 2 requests at a time
                }       
                       )
```

```python
for i in response:
    print(i)
```
