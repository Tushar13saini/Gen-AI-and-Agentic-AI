```python
from pypdf import PdfReader


from langchain_core.tools import tool 

@tool
def pdf_reader_tool(pdf_path: str) -> str:
    """
    Read a PDF file from the given path and return all extracted text.
    Use this tool whenever the user provides a PDF file for summarization,
    question answering, or study purposes.
    """

    try:
        reader = PdfReader(pdf_path)

        text = ""

        for page in reader.pages:
            page_text = page.extract_text()

            if page_text:
                text += page_text + "\n"

        if not text.strip():
            return "No readable text was found in the PDF."

        return text

    except Exception as e:
        return f"Error reading PDF: {e}" 


@tool
def web_scraper_tool(url: str) -> str:
    """Scrape text from an educational webpage."""

    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")

    paragraphs = soup.find_all("p")

    content = "\n".join([p.get_text() for p in paragraphs])

    return content

from langchain_core.messages import HumanMessage


@tool
def notes_summarizer_tool(notes: str) -> str:
    """
    Summarize study notes provided by the user.
    Use this tool whenever the user asks for a summary of notes, a chapter, or PDF content.
    """

    prompt = f"""
    Summarize the following study notes.

    Notes:

    {notes}

    Include:

    1. Important Concepts

    2. Key Definitions

    3. Important Formulae

    4. Revision Points
    """

    response = llm.invoke([HumanMessage(content=prompt)])

    return response.content


@tool
def quiz_generator_tool(
    subject: str,
    chapter: str,
    difficulty: str,
    questions: int
):
    """Generate a quiz based on subject, chapter, difficulty, and number of questions."""

    prompt = f"""
    Subject: {subject}

    Chapter: {chapter}

    Difficulty: {difficulty}

    Generate {questions} MCQs.
    """

    response = llm.invoke(prompt)

    return response.content
```

```python
agent = create_agent( 
    model=llm,
    tools=[pdf_reader_tool, web_scraper_tool, notes_summarizer_tool, quiz_generator_tool],
    system_prompt="""
You are an AI Study Assistant.

Your responsibilities:

- Help students understand difficult concepts.
- Summarize notes accurately.
- Generate quizzes and flashcards.
- Recommend reliable learning resources.
- Use the Web Scraper Tool when additional information is needed.
- Never provide misleading information.
- Explain concepts in simple language.
- Adjust explanations based on the student's difficulty level.

Always return structured answers.
""",
    checkpointer=InMemorySaver()
) 
```

```python
def ask_q(q: str) -> str:
    config = {"configurable": {"thread_id": "session_1"}}
    result = agent.invoke({"messages": [{"role": "user", "content": q}]}, config=config)
    return result["messages"][-1].content 
```
